```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

data = make_classification(n_features=3, n_informative=2, n_redundant=0, n_classes=2, n_clusters_per_class=1, class_sep=2)
X = data[0]
y = data[1]
```

```python
X.shape, y.shape
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)
```

```python
lda_transformed_X = lda.transform(X)
plt.scatter( range(len(y)),lda_transformed_X, c=y)
plt.show()
```
